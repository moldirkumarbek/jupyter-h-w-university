## Q1 (50%)
1. read **data.csv** into Pandas dataframe, and then remove NaN observations (5%)
2. for __Projected Annual Salary__ convert it into float (remove _$_ in front of salary) and delete 0's from dataframe (10%)
3. plot distribution of __Projected Annual Salary__ using histrogram and comment it (10%)
4. is there any outliers in the dataset? answer to that question using appropriate plot (10%)
5. use scatterplot to see the relationship between __Projected Annual Salary__ and some selected variable. Comment their relationship. (15%)

Notes: use only **matplotlib** for visualisation, and of course you can use pandas.

Hint: make sure your plot is properly structured.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Q1.1
data = pd.read_csv("data.csv", sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
data = data.dropna()
data
```

```python
#Q1.2
data_1 = data['Projected Annual Salary'].str.replace('$', '',regex=True).astype(float)
data_1.drop(data_1.index[data_1 == 0], inplace=True)
data_1
```

```python
#Q1.3
plt.figure(figsize=(10,5))
plt.hist(data_1,bins=20)
plt.xlabel('Projected Annual Salary')
plt.ylabel('Frequency')
plt.title("Projected Annual Salary distribution")
plt.show()
```

```python
#Q1.4
plt.figure(figsize=(10,5))
plt.boxplot(data_1,vert=False)
plt.xlabel('Projected Annual Salary')
plt.title("Projected Annual Salary distribution")
plt.show()
```

```python
#Q1.5
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = data['Projected Annual Salary'].str.replace('$', '',regex=True).astype(float), 
           y = data['Employment Type'].str.replace(' ', '_',regex=True))
plt.xlabel("Projected Annual Salary")
plt.ylabel("Employment Type")
plt.show()
# Here we can see the relationship between the length of working hours and wages,
# no matter what kind of work is done for a longer period of time, the higher the salary.
```

```python
ivy_list = ['Brown University', 'Columbia University', 'Cornell University', 'Dartmouth College',
            'Harvard University', 'University of Pennsylvania', 'Princeton University', 'Yale University']
```

```python
# write your code here
import pandas as pd
import matplotlib.pyplot as plt
c_data = pd.read_csv("coursera_data.csv")
c_data.head()
```

```python
#q2.1
c_data1 = c_data.groupby(["course_Certificate_type","course_difficulty"]).size().sort_values(ascending=False)
c_data1.head()
```

```python
#q2.2
c_data2 = c_data[c_data["course_title"].str.contains("Data Science") == True][c_data["course_difficulty"] == 'Intermediate']
c_data2.sort_values(["course_rating"],ascending=False).iloc[0]
```

```python
#q2.3
c_data3 = c_data[c_data["course_organization"].isin(ivy_list)]
c_data3.groupby(["course_organization"]).size().sort_values(ascending=False)
c_data3
```

```python
#q2.4
c_data4 = c_data[c_data["course_organization"] == 'University of Pennsylvania']
c_data4.groupby("course_Certificate_type")["course_rating"]
c_data4.mean()
```

```python
#q2.5
c_data5=c_data.groupby(['course_difficulty'])['course_difficulty'].count().nlargest().reset_index(name="count")
c_data6=c_data.groupby(['course_Certificate_type'])['course_Certificate_type'].count().nlargest().reset_index(name="count")
plt.yticks(rotation=0)
plt.xticks(rotation=270)
plt.bar(c_data5['course_difficulty'],c_data5['count'],color='gold')
plt.bar(c_data6['course_Certificate_type'],c_data6['count'],color='silver')
# The reason why I chose these two is because 
# I want to see how many people here are at what level 
# and how they do it in such a way.
```
