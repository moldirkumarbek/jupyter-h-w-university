```python
#1
with open('words.txt', 'r') as i:
   text= i.read()
print(text)
```

```python
text=text.lower()
print(text) #2
```

```python
words = re.split('[^a-z]',text)
words=[i for i in words if i!=""]
print(words)
```

```python
all_words = [] #4
for i in words:
    i = i.replace('.', '')
    i = i.replace(',', '')
    if i not in all_words and i.islower():
        all_words.append(i)
print(all_words)
```

```python
len(all_words)
```

```python
t = open('words.txt', 'r') #5
text = t.readlines()
text=[i[:-1] for i in text]
statistic=[[i.count(j) for j in all_words] for i in text]
statistic_matrix=np.matrix(statistic)
result = pd.DataFrame(statistic_matrix, index = text, columns=all_words)
result.head()
```

```python
from scipy.spatial.distance import pdist, squareform #6
cos=pd.DataFrame(squareform(pdist(result.values,metric='cosine')), columns=result.index.values,index=result.index.values)
cos.head(1)
```

```python
pd.DataFrame(cos.iloc[0,:]) #second and fifth ar closest to first
```
